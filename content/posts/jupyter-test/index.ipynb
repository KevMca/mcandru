### Visualisations

```python
import numpy as np
import matplotlib.pyplot as plt

time = np.arange(0, 10, 0.1)
amplitude = np.sin(time)

plt.title("Sine save")
plt.plot(time, amplitude)

plt.xlabel("Time")
plt.ylabel("Amplitude")

plt.show()
```
